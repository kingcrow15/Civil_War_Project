```python
data = pd.read_csv('historical_markers.csv')
```

```python
# new data frame with removed ')' within the string
data ['0'] = data["0"].str.split(")", n = 1, expand = True)

data
```

```python
# new data frame with split value columns 
new = data["0"].str.split("(", n = 1, expand = True) 
  
# making separate first name column from new data frame 
data["Marker Name"]= new[0] 
  
# making separate last name column from new data frame 
data["Count"]= new[1] 
  
# Dropping old Name columns 
data.drop(columns =["0", "Unnamed: 0"], inplace = True) 
  
# df display 
data 
```

```python
# data.to_csv("historical_markers_clean.csv")
```

```python
searchfor = ['War', 'Wars']
data["Category"] = data['Marker Name'].apply(lambda x: 'War' if any(i in x for i in searchfor) else 'Non-War')
data
```

```python
data.to_csv("monuments_categorized_.csv")
```
